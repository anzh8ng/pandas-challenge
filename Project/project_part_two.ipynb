```python
# Dependencies
import pandas as pd
```

```python
# load CSV
data_file = "Resources/2016-FCC-New-Coders-Survey-Data.csv"

data_file_df = pd.read_csv(data_file,encoding ='iso-8859-1', low_memory=False) 
data_file_df.head(20)
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
new_df = data_file_df.iloc[:, [0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
new_df.head()
```

```python
new_df.dtypes
```

```python
# Change reading 0 and 1 to No and Yes, respectively
reduced_df = new_df
reduced_df

replace_list = ['HasChildren','AttendedBootcamp','BootcampFinish','BootcampLoanYesNo','BootcampRecommend','BootcampFullJobAfter','CodeEventBootcamp']
reduced_df[replace_list] = reduced_df[replace_list].replace({0:"No",1:"Yes"})

reduced_df.head()



```

```python
# Extract rows for only those who attended a bootcamp
attended_df = reduced_df.loc[reduced_df['AttendedBootcamp']== 'Yes']
attended_df.head()
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
# attended_df=.unique

difbootcamp =pd.DataFrame (attended_df["BootcampName"].value_counts())
difbootcamp.reset_index(inplace=True)
difbootcamp.columns = ["BootcampName", "Number of Attendees"]
difbootcamp.head()
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
recommend_df = attended_df[attended_df["BootcampRecommend"] == "Yes"]["BootcampName"].value_counts()
recommend_df = recommend_df.reset_index().rename(columns = {"index" : "BootcampName", "BootcampName" : "Number of Recommended"})
```

```python
len (attended_df)
```

```python
# Merge the two created data frames on the name of the bootcamp
merged_df = pd.merge(recommend_df,difbootcamp, on="BootcampName", how="right" ).dropna()
merged_df
```

```python
# Calculate percentage of each bootcamp's students who are recommenders
merged_df["%recommend"]=(merged_df["Number of Recommended"]/merged_df["Number of Attendees"])*100 
merged_df

merged_df["%recommend"] = merged_df["%recommend"].map("{0:,.2f}%".format)
merged_df
```

```python
# Export to excel and remove index
export = pd.DataFrame(merged_df)
export.to_excel("grp2_pandapt2.xlsx")
```
