```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# load CSV
csv_path = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas
coder_df = str(pd.read_csv(csv_path)
coder_df.head()
```

```python
# Inspect all columns
coder_df.dtypes
coder_df.describe()
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
reduced_df = coder_df.iloc[:,[0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]

reduced_df.head()
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
       
replace_list = ['HasChildren','AttendedBootcamp','BootcampFinish','BootcampLoanYesNo','BootcampRecommend','BootcampFullJobAfter','CodeEventBootcamp']
reduced_df[replace_list] = reduced_df[replace_list].replace({0:"No",1:"Yes"})

reduced_df.head()

for col in reduced_df:
     print(col, reduced_df[col].unique())
```

```python
# Calculate total number of respondents in survey
total_resp = len(reduced_df)
total_resp
```

```python
# Extract rows corresponding only to people who attended a bootcamp
attended_df = reduced_df.loc[reduced_df['AttendedBootcamp']== 'Yes']
attended_df.head()
```

```python
# Calculate average age of attendees
avg_age = round(attended_df["Age"].mean(),2)

# Calculate how many people attended a bootcamp
attendence = attended_df["AttendedBootcamp"].count()

# Calculate how many attendees hold degrees
#attended_df['SchoolDegree'].value_counts()
degreeslist = ["bachelor's degree","master's degree (non-professional)","professional degree (MBA, MD, JD, etc.)",
           "associate's degree","Ph.D."]

degrees = attended_df['SchoolDegree'].isin(degreeslist).sum()

# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
#attended_df['Gender'].value_counts()
nb_list = ["genderqueer","trans","agender"]

nb_cnt = attended_df['Gender'].isin(nb_list).sum()
m_cnt = attended_df['Gender'].value_counts()['male']
f_cnt = attended_df['Gender'].value_counts()['female']

# Calculate percentage of respondents who attended a bootcamp
percentattend = round((attendence / total_resp) * 100,2)

# Calculate percentage of respondents belonging to each gender
gen_resp = attended_df['Gender'].notna().sum()

m_resp = round((m_cnt/gen_resp)*100,2)
f_resp = round((f_cnt/gen_resp)*100,2)
nb_resp = round((nb_cnt/gen_resp)*100,2)

# Calculate percentage of attendees with a school degree
dg_pct = round((degrees/attendence)*100,2)

# Calculate average post-bootcamp salary
avg_sal = round(attended_df['BootcampPostSalary'].mean(),0)
avg_sal
#print(avg_age,attendence)
```

```python
# Create a new table consolodating above calculations
def as_currency(x):
        return '${:,.2f}'.format(x)
def as_percent(y):
    return "{0:.2f}%".format(y)

table_df = {"Total Surveyed":[total_resp], "Total Bootcamp Attendees":[attendence],
           "% attended bootcamp":as_percent(percentattend), "Avg. Age":[avg_age],"Has a degree":[dg_pct],"% Male":as_percent(m_resp),
           "% Female":as_percent(f_resp),"% Non Gender Specific":as_percent(nb_resp),
            "Average Post Bootcamp Salary": as_currency(avg_sal)}
pd.DataFrame(table_df)
```
